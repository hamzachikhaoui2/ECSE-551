```python
hepatits = pd.read_csv('Hepatits.csv', names = Hepatits_col)
mushroom = pd.read_csv('Mushroom.csv', names = Mushroom_col)
```

```python
fig, axs = plt.subplots(len(hepatits.columns)-1, figsize=(5,20))
fig.subplots_adjust(hspace = 1)
for i, col in enumerate(hepatits.columns[:len(hepatits.columns)-1]):
    axs[i].set_xlabel(f'{col}')
    axs[i].set_ylabel('Occurences')
    axs[i].set_title(f'Distribution of the {col} feature')
    axs[i].hist(hepatits[col],bins = 30)
```
